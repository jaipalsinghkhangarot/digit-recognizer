```python
import pandas as pd
x_data = pd.read_csv("train_digit.csv")
y_data = x_data['label']
x_data = x_data.drop(['label'],axis=1).copy()
x_data
```

```python
test_data = pd.read_csv("test_digit.csv")
test_data
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
train=clf.fit(x_data,y_data)
result=train.predict(test_data)
print(result)
```

```python
Id = []
for i in range(1,28001):
    Id.append(i)
output = pd.DataFrame({'ImageId':Id,'label':result[:]})
output.to_csv('sample_submission_digit.csv')
out = pd.read_csv('sample_submission_digit.csv')
```
